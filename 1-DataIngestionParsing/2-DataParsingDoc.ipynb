Word documents proccessing and handling

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
import os
```

```python
###Method 1: Docx2txtLoader
try:
    docx2txt_loader = Docx2txtLoader("data/word_files/Ai_agent.docx")
    doc = docx2txt_loader.load()
    print(f"Number of documents loaded with Docx2txtLoader: {len(doc)}")
    print(f"content: {doc[0].page_content[:200]}")
    print(f"metadata: {doc[0].metadata}")
    print(doc)

except Exception as e:
    print(f"Error loading DOCX with Docx2txtLoader: {e}")
```

```python
###Method 2: UnstructuredWordDocumentLoader
try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/word_files/Ai_agent.docx", mode="elements")
    unstructured_doc = unstructured_loader.load()
    print(f"Number of documents loaded with UnstructuredWordDocumentLoader: {len(unstructured_doc)}")
    for i, doc in enumerate(unstructured_doc[:3]):
        print(f"Document {i+1}:")
        print(f"type: {doc.metadata.get('category', 'N/A')}")
        print(f"content: {doc.page_content[:100]}")

except Exception as e:
    print(f"Error loading DOCX with UnstructuredWordDocumentLoader: {e}")
```
